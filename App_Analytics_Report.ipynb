# Train, Test data split

```python
X = df_city.copy()
```

```python
X.drop("Conversions", axis = 1, inplace = True)
X
```

```python
Y = df_city['Conversions']
Y
```

Splitting the data

```python
# Splitting the sata
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2, random_state= 10)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
import sklearn.metrics as metrics
y_pred = regressor.predict(X_test)
r2 = metrics.r2_score(y_test,y_pred)
print(r2)
```

```python
df_test = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_test
```

# XG Boost

```python
import numpy as np
import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
```

```python
for lr in [0.001,0.003,0.005,0.008,0.01,0.05,0.08,0.1,0.5,0.7,0.9,1]:
  model = xgb.XGBRegressor(learning_rate = lr, n_estimators = 100, verbosity = 0) # Initialize the model
  model.fit(X_train,y_train) # train the model
  model.score(X_test,y_test) # scoring the model - r2 squared
  xgb_regressor= xgb.XGBRegressor()

  print("Learning rate : ", lr, "Train score: ", model.score(X_train,y_train),"Cross-Val score : ", cross_val_score(xgb_regressor,X_train,y_train, cv=5))
```

```python
model = xgb.XGBRegressor(learning_rate =0.003, n_estimators= 100)
model.fit(X_train,y_train) # train the model
model.score(X_test, y_test) # scoring
```
